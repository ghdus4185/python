## ì‹¤ìŠµë¬¸ì œ - ìŠ¤íƒœí‹±(ì •ì ) ë©”ì†Œë“œ

* `a` ì—°ì‚°ì `b` ì˜ ìˆœì„œë¡œ ì—°ì‚°í•œë‹¤. (`a - b`, `a / b`)
    1. `add(a, b)` : ë§ì…ˆ

````python
# ì•„ë˜ì— ì½”ë“œë¥¼ ì‘ì„±í•´ì£¼ì„¸ìš”.
class Calculator():
    
    @staticmethod
    def add(a, b):
        return a+b
    @staticmethod
    def sub(a, b):
        return a-b
    @staticmethod
    def mul(a, b):
        return a*b
    @staticmethod
    def div(a, b):
        try:
            return a/b
        except:
            return ('0ìœ¼ë¡œ ë‚˜ëˆ„ë©´ ì•ˆë¨')
````

> `Stack` í´ë˜ìŠ¤ë¥¼ ê°„ëµí•˜ê²Œ êµ¬í˜„í•´ë´…ì‹œë‹¤.

> [Stack](https://ko.wikipedia.org/wiki/%EC%8A%A4%ED%83%9D) : ìŠ¤íƒì€ LIFO(Last in First Out)ìœ¼ë¡œ êµ¬ì¡°í™”ëœ ìë£Œêµ¬ì¡°ë¥¼ ëœ»í•©ë‹ˆë‹¤.

ëª¨ë“  ë©”ì„œë“œëŠ” ì¸ìŠ¤í„´ìŠ¤ë©”ì„œë“œ ì…ë‹ˆë‹¤.

1. `empty()`: ìŠ¤íƒì´ ë¹„ì—ˆë‹¤ë©´ ì°¸ì„ ë¦¬í„´í•˜ê³ , ê·¸ë ‡ì§€ ì•Šë‹¤ë©´ ê±°ì§“ì„ ë¦¬í„´í•œë‹¤.

2. `top()`: ìŠ¤íƒì˜ ê°€ì¥ ë§ˆì§€ë§‰ ë°ì´í„°ë¥¼ ë¦¬í„´í•œë‹¤. ìŠ¤íƒì´ ë¹„ì—ˆë‹¤ë©´ Noneì„ ë¦¬í„´í•œë‹¤.

3. `pop()`: ìŠ¤íƒì˜ ê°€ì¥ ë§ˆì§€ë§‰ ë°ì´í„°ì˜ ê°’ì„ ë¦¬í„´í•˜ê³  í•´ë‹¹ ë°ì´í„°ë¥¼ ì‚­ì œí•œë‹¤. ìŠ¤íƒì´ ë¹„ì—ˆë‹¤ë©´ Noneì„ ë¦¬í„´í•œë‹¤.

4. `push()`: ìŠ¤íƒì˜ ê°€ì¥ ë§ˆì§€ë§‰ ë°ì´í„° ë’¤ì— ê°’ì„ ì¶”ê°€í•œë‹¤. ì•„ë¬´ê²ƒë„ ë¦¬í„´í•˜ì§€ ì•ŠëŠ”ë‹¤.

**ë‹¤ ì™„ë£Œí•˜ì‹  ë¶„ë“¤ì€ __repr__ì„ í†µí•´ ì˜ˆì˜ê²Œ ì¶œë ¥ê¹Œì§€ í•´ë´…ì‹œë‹¤.**

````python
# ì—¬ê¸°ì— ì½”ë“œë¥¼ ì‘ì„±í•´ì£¼ì„¸ìš”.
class Stack():
    lifo = []
    
    def __init__(self):
        self.lifo = []
    
    def empty(self):
        if self.lifo == True:
            return False
        else:
            return True

    def top(self):
        if self.lifo:
            return self.lifo[-1]
#         else:
#             return None
            
    def pop(self):
        if not self.empty():
#             return self.lifo.pop()
            last = self.lifo[-1]
            self.lifo = self.lifo[0:-1]
            return last
#         else:
#             return None
    
    def push(self, data):
        self.lifo += [data]

s = Stack()
s2 = Stack()
s3 = Stack()
````

````python
# ì¸ìŠ¤í„´ìŠ¤ë¥¼ í•˜ë‚˜ ë§Œë“¤ê³  ë©”ì†Œë“œ ì¡°ì‘ì„ í•´ë´…ì‹œë‹¤.
s = Stack()
s.empty() # True
s.push(1) # None
s.push(2) # None
s.empty() # False
s.top()   # 2
s.empty() # False
s.pop()   # 2
s.empty() # False
s.top()   # 1
s.pop()   # 1
s.pop()   # None
s.empty() # True
s.push('a')
s.push('b')
s.push('c')
print()
````
